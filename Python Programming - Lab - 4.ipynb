### 07) WAP to print sum of series 1 – 2 + 3 – 4 + 5 – 6 + 7 ... n.

```python
n = int(input("Enter number:"))

sum = 0
for i in range(1,n+1):
    if i%2==0:
        sum -= i
    else:
        sum += i
print(f'Sum: {sum}')
```

### 10) WAP to print GCD of given two numbers.

```python
n = int(input("Enter 1st number:"))
m = int(input("Enter 2nd number:"))

min = n if n<m else m
gcd = 0

for i in range(1,min+1):
    if n%i==0 and m%i==0:
        gcd = i
        
print(f'GCD: {gcd}')
```

### 12) WAP to find whether the given number is Prime or not.

```python
n = int(input("Enter number:"))

count = 0
for i in range(1,n+1):
    if n%i==0:
        count += 1

if count==2:
    print("Number is Prime number.")
else:
    print("Number is Not Prime number.")
```

### 13) WAP to print sum of digits of given number.

```python
n = int(input("Enter number:"))

sum = 0
while n>0:
    rem = n%10
    sum += rem
    n = n // 10
    
print(f'Sum: {sum}')
```

### 14) WAP to check whether the given number is Palindrome or not.

```python
n = int(input("Enter number:"))

res = 0
temp = n
while temp>0:
    rem = temp%10
    res = res*10 + rem
    temp = temp // 10
    
if(res==n):
    print("Palindrome Number")
else:
    print("Not Palindrome Number")
```

### 15) WAP to check whether the given number is an Armstrong Number or not.

```python
n = int(input("Enter number:"))

count = 0
temp = n
while temp>0:
    rem = temp%10
    count += 1
    temp = temp // 10

sum=0
temp = n
while temp>0:
    rem = temp%10
    res = rem**count
    sum += res
    temp = temp // 10

if sum==n:
    print("Armstrong number.")
else:
    print("Not Armstrong number.")
```

### 16) WAP to print all the perfect numbers between 1 to n.

```python
n = int(input("Enter number:"))

sum = 0 
for i in range(1,n):
    if n%i==0:
        sum += i

if sum==n:
    print("Perfect Number")
else:
    print("Not Perfect Number")
```
