### 02) WAP to generate line chart for given data.

```python
x = range(1,11,1)
cxMarks= [8,9,6,3,5,7,4,1,2,6]
cyMarks= [5,8,9,6,3,2,4,8,8,9]
```

```python
plt.plot(x,cxMarks, marker=".", color='g', ls=":", lw=2, label='cxMarks')
plt.legend()
plt.annotate("Max cxMark", xy = [1.9,9.1])
plt.plot(x,cyMarks, marker="X", color='r', ls="-.", lw=2, label='cyMarks')
plt.legend()
plt.annotate("Max cyMark", xy = [10.1,8.9])
plt.title("Result")
plt.xlabel("Marks")
plt.ylabel("Range")
plt.show()
```

### 04) WAP to plot a line chart showing temperature variation during a week. Mark the lowest and highest temperatue.

```python
Days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
Temperature = [20, 32, 31, 33, 34, 43, 36]
```

```python
plt.plot(Days, Temperature, marker="X", color='y', ls="-.", lw=2, label='Temprature')
plt.legend()
plt.title("Weather")
plt.xlabel("Week Days")
plt.ylabel("Temprature")
plt.annotate("Min Temp.", xy = ["Mon", 20])
plt.annotate("Max Temp.", xy = ["Sat", 43])
```

### 05) WAP to create a detailed pie chart representing the monthly expenditure distribution of a family. 
### The chart should display labels, percentage values, title, different colors for each slice, and explode one slice to highlight the highest expense.

```python
Categories = ["Rent", "Food", "Education", "Transport", "Entertainment", "Savings"]
Amount = [15000, 8000, 6000, 3000, 2000, 5000]
```

```python
plt.pie(Amount, labels=Categories, autopct = "%3.2f%%", colors = ['g', 'r', 'purple', 'y', 'pink', 'b'], explode = [0, 0, 0, 0, 0.2, 0])
plt.title("Monthly expenditure")
plt.show()
```
