## Practical 1: Student Marks Analyzer
### Problem Definition
Store student data as a list of tuples. Remove duplicate marks, calculate average marks
and store result in a dictionary.
### Sample Input
students = [
(1, "Amit", [78, 85, 90, 85]),
(2, "Neha", [88, 92, 79, 88]),
(3, "Ravi", [70, 75, 80, 75])
]
### Sample Output
{1: ('Amit', 84.33), 2: ('Neha', 86.33), 3: ('Ravi', 75.0)}

```python
students = [ (1, "Amit", [78, 85, 90, 85]), (2, "Neha", [88, 92, 79, 88]), (3, "Ravi", [70, 75, 80, 75]) ]

result = {}
avg = 0
for i, j, k in students:
    unique_marks = list(set(k))              
    avg = sum(unique_marks) // len(unique_marks) 
    result[i] = (j, avg)
result
```

## Practical 2: Inventory Management
### Problem Definition
Remove out-of-stock items and calculate total inventory value.
### Sample Input
inventory = {
 'pen': (10, 50),
 'book': (100, 0),
 'pencil': (5, 100)
}
### Sample Output
{'pen': (10, 50), 'pencil': (5, 100)}
Total Value: 1000


```python
inventory = { 'pen': (10, 50), 'book': (100, 0), 'pencil': (5, 100) }

res = {}
total = 0
for i,j in inventory.items():
    if j[1]>0:
        res[i] = j
        total = sum(j[0]*j[1] for j in inventory.values())
print(res, "Total Values:", total)
```

## Practical 3: Email Domain Extractor
### Problem Definition
Extract email domains and count occurrences.
### Sample Input
emails = ['a@gmail.com', 'b@yahoo.com', 'c@gmail.com']
### Sample Output
{'gmail.com': 2, 'yahoo.com': 1}

```python
emails = ['a@gmail.com', 'b@yahoo.com', 'c@gmail.com']

Dict = {}
for i in emails:
    d = i.split('@')[1]
    Dict[d] = Dict.get(d, 0) + 1
Dict
```

## Practical 4: Shopping Cart Billing
### Problem Definition
Calculate item-wise cost and total bill.
### Sample Input
cart = [('apple',2), ('banana',3)]
prices = {'apple':50, 'banana':10}
### Sample Output
{'apple': 100, 'banana': 30}
Total Bill: 130


```python
cart = [('apple', 2), ('banana', 3)] 
prices = {'apple': 50, 'banana': 10}

Dict = {}
bill = 0
total = 0
for i in cart:
    for j,v in prices.items():
        if i[0]==j:
            bill = i[1]*v
            total += bill
            Dict.update({j: bill})

print("Total Bill:",total)
Dict
```

## Practical 5: Character Position Index
### Problem Definition
Store positions of each character in a string.
### Sample Input
s = "banana"
### Sample Output
In [14]:
{'b': [0], 'a': [1, 3, 5], 'n': [2, 4]}

```python
s = "banana"
Dict = {}

for i, j in enumerate(s):
    if j not in Dict:
        Dict[j] = []
    Dict[j].append(i)
        
Dict
```

## Practical 6: Survey Data Aggregation
### Problem Definition
Count responses city-wise and choice-wise.
### Sample Input
responses = [
 ('Amit','Rajkot','Yes'),
 ('Neha','Rajkot','No'),
 ('Ravi','Ahmedabad','Yes')
]
### Sample Output
{'Rajkot': {'Yes': 1, 'No': 1}, 'Ahmedabad': {'Yes': 1}}

```python
responses = [ ('Amit','Rajkot','Yes'), ('Neha','Rajkot','No'), ('Ravi','Ahmedabad','Yes') ]
Dict = {}

for i, j, k in responses:
    if j not in Dict:
        Dict[j] = {}
    Dict[j][k] = Dict[j].get(k,0) + 1
Dict
```
