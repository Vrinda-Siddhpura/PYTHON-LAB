### 03) WAP to create Circle class with area and perimeter function to find area and perimeter of circle.

```python
class Circle:
    def __init__(self):
        self.radius = 0

    def getRadius(self):
        self.radius = float(input("Enter Radius:"))
        
    def area(self):
        return 3.14*self.radius*self.radius

    def perimeter(self):
        return 2*3.14*self.radius

c1 = Circle()
c1.getRadius()
print(c1.area())
print(c1.perimeter())
```

### 05) Create a bank account class with methods to deposit, withdraw, and check balance.

```python
class Bank:
    def __init__(self, Account_No, User_Name, Account_Number, balance):
        self.Account_No = Account_No 
        self.User_Name = User_Name
        self.Account_Number = Account_Number
        self.balance = balance
    
    def deposit(self):
        amount = int(input("Enter amount to deposit:"))
        if(amount<=0):
            print("Invalid Amount!!!!")
        self.balance += amount
        
    def withdraw(self):
        amount = int(input("Enter amount to withdraw:"))
        if(amount<=0 or amount>self.balance):
            print("Invalid Amount!!!!")
            return
        self.balance -= amount

    def check(self):
        print("Account No:", self.Account_No)
        print("User Name:", self.User_Name)
        print("Account No:", self.Account_Number)
        print("Balance:", self.balance)

b1 = Bank(101, "Vrinda", 24010101264, 30000)
b1.deposit()
b1.check()
b1.withdraw()
b1.check()
```

### 06) Create a class for managing inventory that includes attributes such as item name, price, quantity, and methods to add, remove, and update items.

```python
class Inventory:
    def __init__(self):
        self.items = {}   
        
    def add_item(self, name, price, quantity):
        if name in self.items:
            self.items[name]['quantity'] += quantity
        else:
            self.items[name] = {
                'price': price,
                'quantity': quantity
            }
        print(f"{name} added/updated successfully.")

    def remove_item(self, name):
        if name in self.items:
            del self.items[name]
            print(f"{name} removed successfully.")
        else:
            print("Item not found.")

    def update_item(self, name, price=None, quantity=None):
        if name in self.items:
            if price is not None:
                self.items[name]['price'] = price
            if quantity is not None:
                self.items[name]['quantity'] = quantity
            print(f"{name} updated successfully.")
        else:
            print("Item not found.")

    def display(self):
        for name, details in self.items.items():
            print(f"Item: {name}, Price: {details['price']}, Quantity: {details['quantity']}")

inv = Inventory()

inv.add_item("Pen", 10, 50)
inv.add_item("Book", 100, 20)

inv.update_item("Pen", quantity=60)

inv.remove_item("Book")

inv.display()
```

### 08) Create one class student_kit
#### Within the student_kit class create one class attribute principal name ( Mr ABC )
#### Create one attendance method and take input as number of days.
#### While creating student take input their name .
#### Create one certificate for each student by taking input of number of days present in class.

```python
class student_kit:
    principal_name = "Mr ABC"

    def __init__(self, name):
        self.name = name

    def attendance(self, days):
        self.days = days

    def certificate(self):
        print("\n----------")
        print("Student Name:", self.name)
        print("Present Days:", self.days)
        print("Principal:", self.principal_name)

n = int(input("Enter Number Of Students:"))
students = []

for i in range(n):
    name = input("Enter Student Name:")
    s = student_kit(name)

    days = int(input("Enter Number Of Present Days:"))
    s.attendance(days)

    students.append(s)

for i in students:
    i.certificate()
```
