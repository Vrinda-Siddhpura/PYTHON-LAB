### 14) WAP to convert degree into Fahrenheit and vice versa.

```python
deg = float(input("Enter degree: "))

f = (deg*(9/5))+32
print(f"Fahrenheit: {f}")

feh = float(input("Enter Fahrenheit: "))

d = (feh-32)*(5/9)
print(f"Degree: {d}")
```

### 15) WAP to get the distance from user into kilometer, and convert it into meter, feet, inches and centimeter.

```python
km = float(input("Enter distance in kilometer: "))

meter = km*1000
centimeter = km*100
feet = km*100000
inches = centimeter/2.54
feet = inches/12

print(f"Kilometer: {km} km")
print(f"meter: {meter: .2f} m")
print(f"feet: {feet: .2f} ft")
print(f"inches: {inches: .2f} in")
print(f"centimeter: {centimeter: .2f} cm")
```
