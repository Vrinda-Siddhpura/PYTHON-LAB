### 07) WAP to accept item prices and calculate total. Raise ValueError if price is negative, otherwise print the total bill amount.

```python
def calculate_total():
    try:
        n = int(input("Enter number of items: "))
        total = 0

        for i in range(n):
            price = float(input(f"Enter price of item {i+1}: "))

            if price < 0:
                raise ValueError("Price cannot be negative!")

            total += price

        print("Total Bill Amount:", total)

    except ValueError as e:
        print("Error:", e)

calculate_total()
```

### 09) Accept 5 subject marks (0-100). Use assert for validation.

#### Sample Input (Error Case)
#####  Enter marks: 110 

#### Sample Output (Error Case)
##### AssertionError: Marks should be between 0 and 100.

#### Sample Input (Normal Case)
##### Valid marks entered

#### Sample Output (Normal Case)
##### Average Marks displayed

```python
try:
    marks = []

    for i in range(5):
        m = int(input("Enter marks: "))
        
        assert 0 <= m <= 100, "Marks should be between 0 and 100."
        
        marks.append(m)

    print("Valid marks entered")
    
    avg = sum(marks) / 5
    print("Average Marks:", avg)

except AssertionError as e:
    print("AssertionError:", e)
```

### 10) WAP that gets password from user. Password must have at least 8 characters and one digit. 
### Raise WeakPasswordError if invalid, print the messsage "Correct" otherwise.

```python
class WeakPasswordError(Exception):
    pass

try:
    password = input("Enter password: ")
    
    if len(password) < 8 or not any(char.isdigit() for char in password):
        raise WeakPasswordError("Password must be at least 8 characters and contain at least one digit.")
    
    print("Correct")

except WeakPasswordError as e:
    print(e)
```
