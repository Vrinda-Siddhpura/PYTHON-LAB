### 01) Write a function to calculate BMI given mass and height. (BMI = mass/h**2)

```python
def BMI(mass, h):
    bmi = mass/h**2
    print("BMI:", bmi)
    
BMI(2,3)
```

### 02) Write a function that add first n numbers.

```python
def add(n):
    sum = 0
    for i in range(1,n+1):
        sum += i
    print("Sum:", sum)

add(5)
```

### 03) Write a function that returns 1 if the given number is Prime or 0 otherwise.

```python
def palindrome(n):
    count = 0
    for i in range(1,n+1):
        if n%i==0:
            count += 1

    if count==2:
        return 1
    else:
        return 0
palindrome(7)
```

### 04) Write a function that returns the list of Prime numbers between given two numbers.

```python
def listOfPrimeNum(n,m):
    li = []
    for i in range(n,m+1):
        count = 0
        for j in range(1,i+1):
            if i%j==0:
                count += 1
                
        if count==2:
            li.append(i)

    print(li)

listOfPrimeNum(1,10)
```

### 05) Write a function that returns True if the given string is Palindrome or False otherwise.

```python
def isPalindrome(s):
    if s==s[::-1]:
        return True
    else:
        return False

isPalindrome("aba")
```

### 06) Write a function that returns the sum of all the elements of the list.

```python
def listSum(li):
    sum = 0
    for i in li:
        sum += i

    print("Sum:", sum)

listSum([1,2,3,4,5])
```

### 07) Write a function to calculate the sum of the first element of each tuples inside the list.

```python
def sumOfFirstEle(li):
    sum = 0
    for i in li:
        sum += i[0]

    print("Sum:", sum)

sumOfFirstEle([(1,2),(3,4),(5,6)])
```

### 08) Write a recursive function to find nth term of Fibonacci Series.

```python
def fibonacciTerm(n):
    if n==0:
        return 0
    elif n==1:
        return 1
    else:
        return fibonacciTerm(n-1) + fibonacciTerm(n-2)

fibonacciTerm(1)
```

### 10) Write a function to get the sum of the scores ending with zero.
#### Example : scores = [200, 456, 300, 100, 234, 678] 
#### Ans = 200 + 300 + 100 = 600

```python
def sumOfEndingZero(li):
    sum = 0
    for i in li:
        if i%10==0:
            sum += i

    print("Sum:", sum)

sumOfEndingZero([200, 456, 300, 100, 234, 678])
```

### 11) Write a function to invert a given Dictionary.
#### hint: keys to values & values to keys
#### Before : {'a': 10, 'b':20, 'c':30, 'd':40}
#### After : {10:'a', 20:'b', 30:'c', 40:'d'}

```python
def invertDict(dict):
    D = {}
    for i,j in dict.items():
        D[j] = i        
    print(D)

invertDict({'a': 10, 'b':20, 'c':30, 'd':40})
```

### 12) Write a function that returns the number of uppercase and lowercase letters in the given string. 
#### example : Input : s1 = AbcDEfgh ,Ouptput : no_upper = 3, no_lower = 5

```python
def countUpperLower(s):
    countUpper = 0
    countLower = 0
    for i in s:
        if i.isupper():
            countUpper += 1
        else:
            countLower += 1

    print("No_Upper:", countUpper, "NO_Lower:", countLower)

countUpperLower("AbcDEfgh")
```

### 14) For the given list of names of students, extract the names having more that 7 characters. Use filter().

```python
def nameOf_7Char(li):
    return list(filter(lambda i : len(i)>7, li))

li = ['ab', 'abc', 'abcdefg', 'abcdefghij', 'abcdefghijklmno']
print(nameOf_7Char(li))
```

### 15) For the given list of names of students, convert the first letter of all the names into uppercase. use map().

```python
def firstCharIntoUpper(li):
    return list(map(lambda i : i.capitalize(), li))

li = ['vrinda', 'aastha', 'shweta', 'khushi']
print(firstCharIntoUpper(li))
```

### 16) Write udfs to call the functions with following types of arguments:
1. Positional Arguments
2. Keyword Arguments
3. Default Arguments
4. Variable Legngth Positional(*args) & variable length Keyword Arguments (**kwargs)
5. Keyword-Only & Positional Only Arguments

```python
# 1.Positional Arguments

def a(a,b):
    return a+b
a(2,3)

# 2.Keyword Arguments

def b(name, age):
    print(f'Hello {name}, Age: {age}')
b(age=19, name='Vrinda')

# 3.Default Arguments

def c(name='Student'):
    print(f'Hello {name}')
c('V')

# 4.Variable Legngth Positional(*args) & variable length Keyword Arguments (**kwargs)

def d(*args, **kwargs):
    print(args, kwargs)
d(1,2,3,4,5, name='Vrinda', age=19)
    
# 5.Keyword-Only & Positional Only Arguments

def e(a,b,c,/):
    print(a, b, c)
e(1,2,3)

def f(*,a,b,c):
    print(a, b, c)
f(a=1,b=2,c=3)
```
