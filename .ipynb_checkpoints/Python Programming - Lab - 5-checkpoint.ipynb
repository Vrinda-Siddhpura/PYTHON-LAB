### 1) Character Category Counter
**Problem:** Count uppercase, lowercase, digits, and special characters in a string.

**Sample Inputs & Outputs:**
- Input: `Python3@Lab` → Uppercase: 2, Lowercase: 7, Digits: 1, Special: 1
- Input: `HELLO123` → Uppercase: 5, Lowercase: 0, Digits: 3, Special: 0
- Input: `hi@#9` → Uppercase: 0, Lowercase: 2, Digits: 1, Special: 2

```python
a = input("Enter String: ")

countUpper = 0
countLower = 0
countDigit = 0
countSpecial = 0
for i in a:
    if i.isupper():
        countUpper += 1
    elif i.islower():
        countLower += 1
    elif i.isdigit():
        countDigit += 1
    else:
        countSpecial += 1

print(f'Uppercase: {countUpper}\nLowercase: {countLower}\nDigits: {countDigit}\nSpecial: {countSpecial}')
```

### 2) Reverse Each Word
- Input: `Python is Easy` → `nohtyP si ysaE`
- Input: `Hello World` → `olleH dlroW`
- Input: `Learn Python Fast` → `nraeL nohtyP tsaF`

```python
a = input("Enter String: ")

li = a.split()
rev = []
for i in li:
    n = i[::-1]
    rev.append(n)
res = " ".join(rev)
print(res)
```

### 4) First and Last Occurrence Finder
- `programming`, `g` → First:3 Last:10
- `banana`, `a` → First:1 Last:5
- `hello`, `z` → Not found

```python
a = input("Enter String: ")
char = input("Enter char to find: ")

pos1 = a.find(char)
pos2 = a.rfind(char)

if (pos1 or pos2)==-1:
    print("Not found")
else:
    print(f'First: {pos1}, Last: {pos2}')
```

### 8) Replace Vowels
- `Programming` → `Pr*gr*mm*ng`
- `Education` → `*d*c*t**n`
- `Sky` → `Sky`

```python
a = input("Enter String: ")
vowel = 'aeiouAEIOU'

for i in a:
    if i in vowel:
        a = a.replace(i, '*')

print(a)
```

### 9) String Compression
- `aaabbcccc` → `a3b2c4`
- `xxxyyyz` → `x3y3z1`
- `abcd` → `a1b1c1d1`

```python
a = input("Enter String: ")

count = 1
s = ""
for i in range(len(a)-1):
    if a[i]==a[i+1]:
        count += 1
    else:
        s += a[i]+str(count)
        count = 1
        
s += a[i+1]+str(count)
s
```

### 10) Toggle Case (without using str.swapcase())
- `PyThOn` → `pYtHoN`
- `HELlo` → `helLO`
- `Python` → `pYTHON`

```python
a = input("Enter String: ")
str = ""
for i in a:
    if i.isupper():
        str += i.lower()
    elif i.islower():
        str += i.upper()
    else:
        str += i
str
```

### 11) Username Validation
#### valid username rules:
1. only alphabets and digits
2. does not start with a digit
3. length ≥ 6

- `User123` → Valid
- `123User` → Invalid
- `Us@12` → Invalid

```python
a = input("Enter Username: ")

if len(a)<6:
    print("Invalid")
elif a[0].isdigit():
    print("Invalid")
elif not a.isalnum():
    print("Invalid")
else:
    print("Valid")
```
