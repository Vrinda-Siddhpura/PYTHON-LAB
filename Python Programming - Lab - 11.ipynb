### Practical 1: Count Lines, Words and Characters

**Problem Definition:**  
Write a Python program to read text from a file and count the total number of lines, words, and characters. The result should be written to an output file.


```python
fp1 = open("Lab11-P1-Read.txt")

fp2 = open("Lab11-P1.txt", "w")
fp2.write("Number Of Lines: " + str(len(fp1.readlines())) + "\n")
fp1.seek(0,0)

li = fp1.read()
data = li.split()
fp2.write("Number Of Words: " + str(len(data)) + "\n")

fp1.seek(0,0)
fp2.write("Number Of Characters: " + str(len(fp1.read())) + "\n")

fp1.close()
fp2.close()
```


### Practical 2: Search a Word in a File

**Problem Definition:**  
Write a Python program to search a given word in a text file and count its occurrences. Store the result in an output file.


```python
fp1 = open("Lab11-P2-Read.txt")
word = input("Enter Word To Search:")
li = fp1.read()
data = li.split()

fp2 = open("Lab11-P2.txt", "w")
count = data.count(word)
fp2.write(str(count))

fp1.close()
fp2.close()
```


### Practical 3: Student Marks and Average

**Problem Definition:**  
Write a Python program to read roll number, name, and marks of students from a file. Store marks in a list, calculate the average using a user-defined function, and write student details along with average marks to an output file.


```python
fp1 = open("Lab11-P3-Read.txt")
fp2 = open("Lab11-P3.txt", "w")

for i in fp1.readlines():
    li=i.split()
    marks = li[2:]
    sum = 0
    count = 0
    for j in marks:
        sum += int(j)
        count += 1
    avg = sum // count
    fp2.write(li[1] + " " + "Avg: " + " " + str(avg) + "\n")

fp1.close()
fp2.close()
```


### Practical 4: Maximum and Minimum using Tuple

**Problem Definition:**  
Write a Python program to read numbers from a file, store them in a tuple, and find the maximum and minimum values. Write the result to an output file.


```python
fp1 = open("Lab11-P4-Read.txt")
fp2 = open("Lab11-P4.txt", "w")

li = fp1.read()
data = list(map(int, li.split()))
tup = tuple(data)
fp2.write("Maximum: " + str(max(tup)) + "\n")
fp2.write("Maximum: " + str(min(tup)) + "\n")

fp1.close()
fp2.close()
```


### Practical 5: Remove Duplicate Words using Set

**Problem Definition:**  
Write a Python program to read words from a file and remove duplicate words using a set. Store the unique words in an output file.


```python
fp1 = open("Lab11-P5-Read.txt" , "r")
fp2 = open("Lab11-P5.txt", "w")

li = fp1.read()
data = li.split()
s = set(data)
fp2.write("Unique Words: " + ",".join(s))

fp1.close()
fp2.close()
```


### Practical 6: Word Frequency using Dictionary

**Problem Definition:**  
Write a Python program to count the frequency of each word in a text file using a dictionary and write the result to an output file.


```python
fp1 = open("Lab11-P6-Read.txt")
fp2 = open("Lab11-P6.txt", "w")
li = fp1.read()
data = li.split()

Dict = {}
for i in data:
    c = data.count(i)
    Dict.update({i: c})
fp2.write(str(Dict))

fp1.close()
fp2.close()
```


### Practical 7: Student Result and Grade Calculation

**Problem Definition:**  
Write a Python program to read roll number, name, and marks of students from a file. Use a user-defined function to determine grade (A/B/C/Fail) and write the student result to an output file.

**Grade Criteria:**  
- Marks ≥ 75 : Grade A  
- Marks ≥ 60 : Grade B  
- Marks ≥ 40 : Grade C  
- Marks < 40 : Fail


```python
fp1 = open("Lab11-P7.txt", "w")

rollNo = int(input("Enter Roll_No.:"))
name = input("Enter Name:")
mark = int(input("Enter Mark:"))

def result(mark):
    if mark>=75:
        return "A"
    elif mark>=60 and mark<75:
        return "B"
    elif mark>=40 and mark<60:
        return "C"
    else:
        return "Fail"

grade = result(mark)

fp1.write("Roll_No: " + str(rollNo) + " " + "Name: " + name + " " + "Mark: " + str(mark) + "\n" + "Result: " + grade)

fp1.close()
```
